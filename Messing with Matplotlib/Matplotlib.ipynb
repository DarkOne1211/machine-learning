```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
```

```python
y = x**2
```

```python
x
y
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'g')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes2.plot(y,x)
axes2.set_title('Smaller plot')
axes1.set_title('larger plot')
```

```python
fig,axes = plt.subplots(1,2)
for current_ax in axes: 
    current_ax.plot(x,y)
```

```python
fig
```

```python
fig.savefig('my_picture.png',dpi = 200)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='x ^ 2')
ax.set_title('Title')
ax.set_ylabel('Y label')
ax.set_xlabel('X label')
ax.plot(x,x**3, label = 'x ^ 3')
ax.legend()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='blue', linewidth = 2, linestyle='-')
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
